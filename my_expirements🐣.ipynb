```python
import cv2
import matplotlib.pyplot as plt

from generator.augmentation import augment_img
```

```python
image = cv2.imread("demo_image_cropped.png")
plt.imshow(image)
None
```

```python
fig, axes = plt.subplots(4, 4, figsize=(8, 6))

axes[0, 0].imshow(image)
axes[0, 0].axis('off')

for i in range(1, 16):
    augmented_image = augment_img(image)
    axes[i // 4, i % 4].imshow(augmented_image)
    axes[i // 4, i % 4].axis('off')

plt.tight_layout()
plt.show()
```
