```python
from generator.generator import Generator
from generator.font import *
import matplotlib.pyplot as plt
import pandas as pd
from tqdm.contrib import tzip

tatneft = pd.read_csv("generated_dataset/labels/labels.csv")
g = Generator(lang='ru')
```

```python
files = []
names = []
for n, name in tzip(range(len(tatneft)), tatneft['name']):

    s = g.generate_from_string(name, min_length=2, max_length=25)
    if s:
        file = "gen_" + str(n) + ".jpg"
        files.append(file)
        names.append(name)
        s.convert('RGB').save(f"generated_dataset/images/{file}")

pd.DataFrame({"file": files, "name": names}).to_csv("generated_dataset/labels/labels.csv", index=False)
```
