```python
from generator.generator import Generator
from generator.font import *
import matplotlib.pyplot as plt
import numpy as np

g = Generator(lang='ru')
plt_dim = int(np.ceil(np.sqrt(g.N)))
fig, axs = plt.subplots(plt_dim, plt_dim, figsize=(20, 20))

for k in range(plt_dim):
    for j in range(plt_dim):
        s = g.generate_from_string(r'Площадь:', min_length=2, max_length=25, i = k * plt_dim + j) # get from a string
        if s:
            font_name = g.fonts[k * plt_dim + j].path
            axs[k, j].set_title(font_name[font_name.rfind('\\')+1:])
            axs[k, j].imshow(s)
plt.show()
```
